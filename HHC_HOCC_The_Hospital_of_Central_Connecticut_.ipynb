## Length of stay

```python
tmp = connecticut[['PAT_MRN_ID','PAT_ENC_CSN_ID','EFFECTIVE_DTTM','EVENT_TYPE']]
adm_con = tmp[tmp['EVENT_TYPE'] == 'Admission'].reset_index(drop=True)
dis_con = tmp[tmp['EVENT_TYPE'] == 'Discharge'].reset_index(drop=True)
adm_dis_date = dis_con.join(adm_con.set_index('PAT_ENC_CSN_ID'),on = 'PAT_ENC_CSN_ID', lsuffix='_x',rsuffix='_y')
```

```python
adm_dis_date = adm_dis_date.loc[adm_dis_date['EFFECTIVE_DTTM_y'].isnull() == False] 

adm_date_list = []
discharge_date_list = []
los = []

for i in range(len(adm_dis_date)):
  adt_date = datetime.strptime(adm_dis_date['EFFECTIVE_DTTM_y'].iloc[i], '%m-%d-%Y %H:%M')
  dis_date = datetime.strptime(adm_dis_date['EFFECTIVE_DTTM_x'].iloc[i], '%m-%d-%Y %H:%M')
  adm_date_list.append(adt_date)
  discharge_date_list.append(dis_date)
  date_diff = abs(adt_date - dis_date).days
  los.append(date_diff)

adm_dis_date['adm_date'] = adm_date_list
adm_dis_date['dis_date'] = discharge_date_list
adm_dis_date['los'] = los
```

```python
pd.unique(adm_dis_date['los'])
```

```python
agg_los = adm_dis_date.groupby('los').size().reset_index()
cnt_list = [0]*5

for i in range(len(agg_los)):
  los = agg_los['los'].iloc[i]
  cnt = agg_los[0].iloc[i]
  if los == 0:
    cnt_list[0] += cnt
  elif los == 1:
    cnt_list[1] += cnt
  elif 1<los<=7:
    cnt_list[2] += cnt
  elif 7<los<=31:
    cnt_list[3] += cnt
  else:
    cnt_list[4] += cnt
```

```python
tmp = adm_dis_date.groupby('los').size().reset_index()
x=cnt_list
labels = ['Same day', '<= 1day', '<= 1week', '<= 1month', '> 1month']
fig, ax = plt.subplots()
ax.pie(x, labels=labels,autopct='%1.1f%%')
ax.set_title('Pie Chart for Length of Stay (all data)')
fig.set_size_inches(7,7)
plt.tight_layout()
```

```python
tmp2 = tmp.iloc[1: , :]
x=cnt_list[1:]
labels = ['<= 1day', '<= 1week', '<= 1month', '> 1month']
fig, ax = plt.subplots()
ax.pie(x, labels=labels,autopct='%1.1f%%')
ax.set_title('Pie Chart for Length of Stay (Same Day Discharge Excluded))')
fig.set_size_inches(7,7)
plt.tight_layout()
```

## Discharge/admission in sept

```python
this_date = []

for i in range(len(connecticut)):
  tmp = datetime.strptime(connecticut['EFFECTIVE_DTTM'].iloc[i], '%m-%d-%Y %H:%M')
  tmp = tmp.month
  this_date.append(tmp)
connecticut['date_month'] = this_date
```

```python
tmp9 = connecticut[connecticut['date_month'] == 9]
groupby9 = tmp9.groupby(['DEPARTMENT_NAME','EVENT_TYPE']).size().reset_index()
groupby9 = groupby9[groupby9['EVENT_TYPE'] == 'Admission']
groupby9 = groupby9.sort_values(by=[0],ascending=False).reset_index(drop=True)
groupby9 = groupby9.rename(columns={0:'COUNTS'})
groupby9
```

```python
menMeans = groupby9['COUNTS'][:10]
ind = groupby9['DEPARTMENT_NAME'][:10]
y_pos = range(len(ind))

#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (10,8))
ax.bar(ind,menMeans,width=0.4)
plt.xticks(y_pos, ind, rotation=30)
for index,data in enumerate(menMeans):
    plt.text(x=index-0.25 , y =data+200 , s=f"{data}" , fontdict=dict(fontsize=10))
plt.tight_layout()
ax.set_title("Top-10 Admission Count Departments")
ax.set_xlabel("Department Name")
ax.set_ylabel("Admission Count")
plt.show()
```
